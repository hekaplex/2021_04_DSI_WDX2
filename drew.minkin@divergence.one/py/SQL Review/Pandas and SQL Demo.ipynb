**AS or derived column review**

```python
# WITH 
#     NVVar1_Std as
#         (
#             SELECT
#                 STDEVP(NVVar1) AS NVVar1_Std
#             FROM 
#                 df
#         )
#     NVVar1_Avg as
#         (
#             SELECT
#                 AVG(NVVar1) AS NVVar1_Avg
#             FROM 
#                 df
#         )
# SELECT
#     Calendar_Year
#     ,NVVar1
#     ,
#         (
#             NVVar1
#             -
#             NVVar1_Avg.NVVar1_Avg
#         )
#         /
#         NVVar1_Std.NVVar1_Std
#     AS NVVar1_Z_Score
# FROM
#     df
# FULL OUTER JOIN 
#     NVVar1_Std
# ON
#     1=1
# FULL OUTER JOIN 
#     NVVar1_Avg
# ON
#     1=1
df['NVVar1_ZScore'] \
    = (
            df['NVVar1']  \
            -  \
            df['NVVar1'].mean() \
        )  \
            /  \
        df['NVVar1'].std(ddof=0)

df['NVVar1_ZScore']     = (df['NVVar1'] -  df['NVVar1'].mean() )  /df['NVVar1'].std(ddof=0)
df['NVVar1_ZScore2']= stats.zscore(df['NVVar1'], axis=0, ddof=1).to_frame()
df[["Calendar_Year","NVVar1","NVVar1_ZScore","NVVar1_ZScore2"]]
```
